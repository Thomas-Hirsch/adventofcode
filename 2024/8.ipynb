```python
from aocd import get_data
from collections import defaultdict
import itertools

grid = get_data(year=2024,day=8).splitlines()

def in_bounds(x,y,grid):
    return 0 <= x < len(grid[0]) and 0 <= y < len(grid)

all_antinodes = set()
antennae = defaultdict(list)

for y in range(len(grid)):
    for x in range(len(grid[0])):
        if grid[y][x] != ".":
            antennae[grid[y][x]].append((x,y))

def get_antinodes(a,b,grid):
    xdiff = a[0] - b[0]
    ydiff = a[1] - b[1]

    first = (a[0] + xdiff, a[1] + ydiff)
    second = (b[0] - xdiff, b[1] - ydiff)
    antinodes = [x for x in [first, second] if in_bounds(x[0],x[1],grid)]
    return antinodes


for ant in antennae:
    node_pairs = itertools.combinations(antennae[ant],2)
    antinodes = [get_antinodes(p[0],p[1],grid) for p in node_pairs]
    tidied = [x for l in antinodes for x in l]
    for t in tidied:
        all_antinodes.add(t)

def part2(a,b,grid):
    dx = a[0] - b[0]
    dy = a[1] - b[1]
    
    antinodes = set()
    row, col = a
    while in_bounds(row,col,grid):
        antinodes.add((row,col))
        row += dx
        col += dy
    
    row, col = b
    while in_bounds(row,col,grid):
        antinodes.add((row,col))
        row -= dx
        col -= dy
    return antinodes

antinodes_2 = set()
for ant in antennae:
    node_pairs = itertools.combinations(antennae[ant],2)
    antinodes = [part2(p[0],p[1],grid) for p in node_pairs]
    tidied = [x for l in antinodes for x in l]
    for t in tidied:
        antinodes_2.add(t)


print(len(antinodes_2))
```
