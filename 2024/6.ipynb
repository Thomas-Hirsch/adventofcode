```python
%%time
from aocd import get_data

data = [list(line) for line in get_data(year=2024,day=6).splitlines()]

def in_bounds(x,y,grid):
    return 0 <= x < len(grid[0]) and 0 <= y < len(grid)

def get_start(grid):
    for y, line in enumerate(grid):
        for x, v in enumerate(line):
            if v == "^":
                return (x,y)
            
start = get_start(data)
orientation = 0
visited = set()
visited.add(start)
directions = [(0,-1), (1,0), (0,1), (-1,0)]
x,y = start
while in_bounds(x,y,data):
    visited.add((x,y))
    direction = orientation % 4
    next_x, next_y = x + directions[direction][0], y + directions[direction][1]
    if data[next_y][next_x] == "#":
        orientation += 1
    else:
        x, y = next_x, next_y
len(visited)
```

```python
obstacles = 0
iterations = 0
for x,y in list(visited):
    iterations += 1
    if iterations % 10 == 0:
        print(f"iteration: {iterations}")
    orientation = 0
    temp_grid = data.copy()
    temp_grid[y][x] = "#"
    temp_visited = set()
    x,y = get_start(data)
    while(in_bounds(x,y,temp_grid)):
        direction = orientation % 4
        if (x,y,orientation) in temp_visited:
            obstacles += 1
            break
        else:
            next_x, next_y = x + directions[direction][0], y + directions[direction][1]
            if  in_bounds(next_x,next_y,temp_grid) and temp_grid[next_y][next_x] == "#":
                orientation += 1
            else:
                x, y = next_x, next_y

obstacles
```
